## Calculating the Return of Indices

Consider three famous American market indices â€“ Dow Jones, S&P 500, and the Nasdaq for the period of 1st of January 2019 until today.

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['^DJI', '^GSPC', '^IXIC']

ind_data = pd.DataFrame()

for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo', start='2019-1-1')['Adj Close']
```

```python
ind_data.head()
```

```python
ind_data.tail()
```

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(15, 6));
plt.show()
```

```python
# calculating rate of return of major indices in USA
ind_returns = (ind_data / ind_data.shift(1)) - 1

ind_returns.tail()
```

Estimate the average annual return of each index.

```python
annual_ind_returns = ind_returns.mean() * 250
annual_ind_returns
```
